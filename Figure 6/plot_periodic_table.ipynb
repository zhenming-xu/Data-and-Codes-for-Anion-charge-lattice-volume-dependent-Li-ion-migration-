```python
import matplotlib.pyplot as plt
from periodic_table import periodic_table
%cd ~/Desktop/Charge-Barrier/plot_periodic_table
```

```python
plt.close()
p = periodic_table(
    frame={
        "shape": "Rectangle",
        "parms": [1.0, 1.2]
    },
    labels={
        "background": {
            "Six-Coordinated Non-Mobile Cation Elements": ['Sc', 'Y', 'Zr', 'Hf', 'La', 'Ce', 'Pr', 'Nd'],
            "Four-Coordinated Non-Mobile Cation Elements": ['N', 'Si', 'Ge', 'P', 'As', 'Sn', 'Sb'],
            "Mobile Cation Elements": ['Li','Na', 'K'],
            "Anion Elements": ['O', 'S', 'Se', 'Te', 'F', 'Cl', 'Br', 'I']
        }
    })
pplt = p.get_plot(figsize=(12, 6))
pplt.savefig("periodic_table-1.pdf")
pplt.savefig("periodic_table-1.png", dpi=600)
pplt.show()
```
